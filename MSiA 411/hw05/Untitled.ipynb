```python
import pandas as pd
import datetime as dt
```

```python
df = pd.read_excel("MSiA_Assignment 4_dataset-1.xlsx")
```

```python
df['counts'] = df.groupby('Name of charity/Project')['Name of charity/Project'].transform('count')
```

```python
df_2 = df.groupby('Charity Country').count().reset_index()[['Charity Country', 'counts']]
```

```python
df_3 = df.groupby('Volunteer Country').count().reset_index()[['Volunteer Country', 'counts']]
```

```python
df_2.to_excel("v_counts_data.xlsx")
```

```python
def get_year(row):
    return row.year
```

```python
df['year_of_proj'] = df['Date of project'].apply(get_year)
```

```python
df_4_non_us = df[df['Charity Country']!='United States'].groupby('year_of_proj').count().reset_index()[['year_of_proj', 'counts']]
df_4_us = df[df['Charity Country']=='United States'].groupby('year_of_proj').count().reset_index()[['year_of_proj', 'counts']]
df_4_us = df_4_us.rename(columns={'year_of_proj': "year_of_proj_us", 'counts': "counts_us"})
```

```python
df_4 = pd.concat([df_4_non_us, df_4_us], axis=1)
```

```python
df_4
```

```python
df_4['Projects Outside of U.S.'] = df_4['counts'].cumsum()
df_4['Projects Inside of U.S.'] = df_4['counts_us'].cumsum()
df_4
```

```python
df_4.to_excel('df_4.xlsx')
```
